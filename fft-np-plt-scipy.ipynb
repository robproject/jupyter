```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack

# Sample points
N = 1000
# sample spacing
T = 1 
x = np.linspace(0, 1000, 1)

def plt_fft(y):

    yf = scipy.fftpack.fft(y)
    xf = np.linspace(0, 1000, 1)

    fig, ax = plt.subplots()
    ax.plot(xf, 2.0/N * np.abs(yf[:N//2]))
    plt.grid(visible=1)
    plt.show()

y = np.cos(10000 * np.pi * x)
plt_fft(y)

#y = np.sin(50 * 2 * np.pi * x) + 0.5 * np.sin(80 * 2 * np.pi * x)
#plt_fft(y)




```
